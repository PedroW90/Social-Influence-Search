```python
c_degree = nx.degree_centrality(G_fb)
c_degree = list(c_degree.values())
```

```python
c_eigenvector = nx.eigenvector_centrality(G_fb)
c_eigenvector = list(c_eigenvector.values())
```

```python
c_betweenness = nx.betweenness_centrality(G_fb)
c_betweenness = list(c_betweenness.values())
```

```python
import time
from matplotlib import pyplot as plt
```

```python
from networkx.algorithms import community
```

```python

#partition = community.best_partition(G_fb)
pos = nx.spring_layout(G_fb)
plt.figure(figsize=(8, 8))
plt.axis('off')
nx.draw_networkx_nodes(G_fb, pos,cmap=plt.cm.RdYlBu)
nx.draw_networkx_edges(G_fb, pos, alpha=0.3)
plt.show(G_fb)
```

```python
# Plot the centrality of the nodes
tic=time.time()
plt.figure(figsize=(18, 12))# Degree Centrality
f, axarr = plt.subplots(2, 2, num=1)
plt.sca(axarr[0,0])
nx.draw(G_fb, cmap = plt.get_cmap('inferno'),pos=pos, node_color = c_degree)
axarr[0,0].set_title('Degree Centrality', size=16)# Eigenvalue Centrality
plt.sca(axarr[0,1])
nx.draw(G_fb, cmap = plt.get_cmap('inferno'), pos=pos, node_color = c_eigenvector)
axarr[0,1].set_title('Eigenvalue Centrality', size=16)# Proximity Centrality
plt.sca(axarr[1,0])
nx.draw(G_fb, cmap = plt.get_cmap('inferno'), pos=pos,node_color = c_closeness)
axarr[1,0].set_title('Proximity Centrality', size=16)# Betweenness Centrality
plt.sca(axarr[1,1])
nx.draw(G_fb, cmap = plt.get_cmap('inferno'),pos=pos, node_color = c_betweenness)
axarr[1,1].set_title('Betweenness Centrality', size=16)

toc=time.time()


print(tic-toc)
```
